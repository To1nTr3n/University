```python
# Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, sdt, Q1, 
# Q2 và Q3 của DH1 theo KT
#pd.pivot_table(df, index= ["KT"], values= ['DH1'], aggfunc=['count',np.sum, np.mean,np.median, np.std, np.min, np.max])
#3
#df.groupby('KT')['DH1'].agg(['count','sum','mean','median','min','max','std'])
df_KT = df.groupby('KT')['DH1'].aggregate(['count',np.sum, np.mean,np.median, np.std, np.min, np.max, lambda x: x.quantile(0.25) ,lambda x: x.quantile(0.5) ,lambda x: x.quantile(0.75)])
df_KT = df_KT.rename(columns = {'<lambda_0>':'Q1','<lambda_1>':'Q2','<lambda_2>':'Q3'})
#pd.pivot_table(df_KT, columns='KT')
df_KT
```

```python
#4
#df.groupby(['KT','KV'])['DH1'].agg(['count','sum','mean','median','min','max','std'])
#pd.pivot_table(df, index= ["KT",'KV'], values= ['DH1'], aggfunc=[np.sum, np.mean,np.median, np.std, np.min, np.max, 'count'])
df_KT_KV = df.groupby(['KT','KV'])['DH1'].aggregate(['count',np.sum, np.mean,np.median, np.std, np.min, np.max, lambda x: x.quantile(0.25) ,lambda x: x.quantile(0.5) ,lambda x: x.quantile(0.75)])
df_KT_KV = df_KT_KV.rename(columns = {'<lambda_0>':'Q1','<lambda_1>':'Q2','<lambda_2>':'Q3'})
# pd.pivot_table(df_KT_KV)
df_KT_KV
```

# Phần 3: Trực quan hóa dữ liệu theo nhóm phân loại
## 1. Trực quan dữ liệu học sinh nữ trên các nhóm XL1, XL2, XL3 dạng unstacked


```python
# Lọc dữ liệu giới tính là nữ
# - Oy: Chiều cao biểu đồ cột thể hiển số lượng học sinh theo xếp loại
# - Màu sắc thể hiện giá trị xếp loại: [Y, TB, K, G, XS]
# - Ox: thể hiện nhóm XL1, XL2 và XL3
```

```python
df_nu_xl1 = df[df.GT == 'F'].groupby(['XL1'])['XL1'].agg(['count']).transpose()
df_nu_xl2 = df[df.GT == 'F'].groupby(['XL2'])['XL2'].agg(['count']).transpose()
df_nu_xl3 = df[df.GT == 'F'].groupby(['XL3'])['XL3'].agg(['count']).transpose()
#df_nu_xl1.plot.bar(rot=0);#df_nu_xl1.plot.pie(y = 'count', subplots=True);df_nu_xl1
index = ["Y","TB", "K", "G", "XS"]
df_nu_xl1,df_nu_xl2,df_nu_xl3
```

```python
xl = ["XL1","XL2","XL3"]
df_xl = pd.concat([df_nu_xl1, df_nu_xl2, df_nu_xl3], ignore_index = True )
df_xl['xl'] = ["XL1","XL2","XL3"]
df_xl['XS'] = [0,0,0]
df_xl = df_xl.fillna(0)
df_xl
```

```python
df_xl = df_xl.set_index('xl')
df_xl
```

```python
df_xl.plot.bar()
```

### 2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:

    a. Từ 0 đến dưới 5 = kém (ký hiệu "k")
    
    b. Từ 5 đến dưới 7 = trung bình (ký hiệu "tb")
    
    c. Từ 7 đến dưới 8 = khá (ký hiệu "kh")
    
    d. Từ 8 trở lên = giỏi (ký hiệu "g")

```python
for i in df['T1']:
    df.loc[(df['T1'] >= 0) & (df['T1'] < 5), 'phanlopt1'] = 'Kem'
    df.loc[(df['T1'] >= 5) & (df['T1'] < 7), 'phanlopt1'] = 'TB'
    df.loc[(df['T1'] >= 7) & (df['T1'] < 8), 'phanlopt1'] = 'Kha'
    df.loc[(df['T1'] >= 8), 'phanlopt1'] = 'Gioi'
```

```python
df.phanlopt1
```

```python
df_plT1 = df.groupby('phanlopt1')['phanlopt1'].agg(['count'])
df_plT1
```
